```python
import random as rd
import pandas as pd
import os

print("Geração média anual esperada: 15.000 kWh")

geracao_mensal = 0

meses = {"jan":1,"fev":2,"mar":3,
         "abr":4,"mai":5,"jun":6,
         "jul":7,"ago":8,"set":9,
         "out":10,"nov":11,"dez":12}

alertas_log = {}

eventos_ponderados = [
    1,      #Coeficiente 1 com 40% de chance
    0.8,    # Coeficiente 0.8 com 20% de chance
    0.4,    # Coeficiente 0.4 com 20% de chance
    0.2,    # Coeficiente 0.2 com 15% de chance
    0.0     # Coeficiente 0.0 com 05% de chance
]
pesos = [0.4, 
         0.20,
         0.20, 
         0.15, 
         0.05]

geracao_anual = []
eficiencia_anual = []
total_alertas = 0
baixa_producao = 0
media_geracao = 0
arquivo_existe = os.path.isfile('arquivo.csv')



for mes_nome, mes_numero in meses.items():
    if mes_numero in [12, 1, 2]:
        for dia in range(1, 30):

            eventos_aleatorios = rd.choices(eventos_ponderados,weights=pesos)       
            aleatoriedade = eventos_aleatorios[0]

            if aleatoriedade == 0:
                total_alertas += 1

                if mes_nome not in alertas_log:
                    alertas_log[mes_nome] = 1
                else:
                    alertas_log[mes_nome] += 1


            eficiencia = rd.uniform(0.9, 0.95)
            geracao_base_dia = 1000 * eficiencia * aleatoriedade
            geracao_mensal += geracao_base_dia
            eficiencia_anual.append(eficiencia)

        if geracao_mensal <= 15000:
            baixa_producao += 1
        
        print(f"Geração em {mes_nome}: {geracao_mensal:.0f} kWh")
        geracao_anual.append(geracao_mensal)
        geracao_mensal = 0          # Reinicie a geração mensal para o próximo mês.

    elif mes_numero in [3, 4, 5]:
        for dia in range(1, 30):

            eventos_aleatorios = rd.choices(eventos_ponderados,weights=pesos)       
            aleatoriedade = eventos_aleatorios[0]

            if aleatoriedade == 0:
                total_alertas += 1

                if mes_nome not in alertas_log:
                    alertas_log[mes_nome] = 1
                else:
                    alertas_log[mes_nome] += 1


            eficiencia = rd.uniform(0.75, 0.85)
            geracao_base_dia = 1000 * eficiencia * aleatoriedade
            geracao_mensal += geracao_base_dia
            eficiencia_anual.append(eficiencia)


        if geracao_mensal < 15000:
            baixa_producao += 1
        
        print(f"Geração em {mes_nome}: {geracao_mensal:.0f} kWh")
        geracao_anual.append(geracao_mensal)
        geracao_mensal = 0

    elif mes_numero in [6, 7, 8]:
        for dia in range(1, 30):

            eventos_aleatorios = rd.choices(eventos_ponderados,weights=pesos)       
            aleatoriedade = eventos_aleatorios[0]

            if aleatoriedade == 0:
                total_alertas += 1

                if mes_nome not in alertas_log:
                    alertas_log[mes_nome] = 1
                else:
                    alertas_log[mes_nome] += 1


            eficiencia = rd.uniform(0.6, 0.7)
            geracao_base_dia = 1000 * eficiencia * aleatoriedade
            geracao_mensal += geracao_base_dia
            eficiencia_anual.append(eficiencia)

        if geracao_mensal < 15000:
            baixa_producao += 1

        print(f"Geração em {mes_nome}: {geracao_mensal:.0f} kWh")
        geracao_anual.append(geracao_mensal)
        geracao_mensal = 0

    elif mes_numero in [9, 10, 11]:
        for dia in range(1, 30):

            eventos_aleatorios = rd.choices(eventos_ponderados,weights=pesos)       
            aleatoriedade = eventos_aleatorios[0]

            if aleatoriedade == 0:
                total_alertas += 1

                if mes_nome not in alertas_log:
                    alertas_log[mes_nome] = 1
                else:
                    alertas_log[mes_nome] += 1


            eficiencia = rd.uniform(0.75, 0.85)
            geracao_base_dia = 1000 * eficiencia * aleatoriedade
            geracao_mensal += geracao_base_dia
            eficiencia_anual.append(eficiencia)

        if geracao_mensal < 15000:
            baixa_producao += 1

        print(f"Geração em {mes_nome}: {geracao_mensal:.0f} kWh")
        geracao_anual.append(geracao_mensal)
        geracao_mensal = 0


soma_geracao = sum(geracao_anual)
media_geracao_anual = soma_geracao / 12
media_eficiencia = sum(eficiencia_anual)/len(eficiencia_anual)


print("Log de alertas / falhas:",alertas_log)
print("Total de falhas / geração comprometida:",total_alertas)
print(f"Media: {media_geracao_anual:.0f} kWh")

#Vamos considerar que a geração esperada seja de 15.000 kWh

if media_geracao_anual < 15000:
    meta_anual = 'Sim'
else:
    meta_anual = 'Não'



dados = {
    'Media Geracao Anual': [round(media_geracao_anual, 2)],
    'Eficiencia media':[round(media_eficiencia, 3)],
    'Meses com Producao Abaixo do Esperado': [baixa_producao],
    'Dias com Falha / Geração Comprometida': [total_alertas],
    'Media Anual abaixo':[meta_anual]
}

df = pd.DataFrame(dados)
excel_file = 'tabela_energia_solar.xlsx'

if not arquivo_existe:
    df.to_csv('arquivo.csv',mode='a', index=False)
else:
    df.to_csv('arquivo.csv',mode='a',header=False, index=False)

df = pd.read_csv('arquivo.csv',index_col=0)

df.to_excel(excel_file)

# print(data)
```
